```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
def denoise_image(image_path):
    image = plt.imread(image_path)
    denoised_image = cv2.fastNlMeansDenoisingColored(image, None, 13, 10, 7, 21)
    img = np.concatenate((image,denoised_image),axis=1)
    
    plt.figure(figsize=(15,15))
    plt.imshow(img)
    image_path_new = 'denoised_images/' + image_path
    plt.imsave(image_path_new , img)
    return True
```

```python
denoise_image('human_image.jpg')
```
