```python
from PIL import Image,ImageFilter, ImageEnhance 
import numpy as np 
import cv2 
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
def image_enhance(image_path):
    image = Image.open(image_path)
    enhancer = ImageEnhance.Sharpness(image)
    #enhancer= ImageEnhance.Contrast(image) 
    enhanced_im = enhancer.enhance(1.8)
    #enhanced_im.save(image_path)
    return enhanced_im
```

```python
def image_enhance_smooth(image_path):
    image = Image.open(image_path)
    im = image.filter(ImageFilter.BLUR)
    enhancer = ImageEnhance.Sharpness(im)
    enhancer= ImageEnhance.Contrast(im)  
    enhanced_im = enhancer.enhance(1.8)
    #enhanced_im.save(image_path)
    return enhanced_im
```

```python
image_enhance_smooth("Tanush Pas/PSX_20180521_183644.jpg")
```

```python
def image_enhance_minfilter(image_path):
    image = Image.open(image_path)
    im = image.filter(ImageFilter.MinFilter(3))
    enhancer = ImageEnhance.Sharpness(im)
    enhancer= ImageEnhance.Contrast(im)  
    enhanced_im = enhancer.enhance(1.8)
    #enhanced_im.save(image_path)
    return enhanced_im
```

```python
image_enhance_minfilter("Tanush Pas/PSX_20180521_183644.jpg")
```
